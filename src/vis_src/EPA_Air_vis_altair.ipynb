```python
import altair as alt
import pandas as pd

co_air_df = pd.read_csv("../../data/final/epa_air_params/co_2020_avg_df.csv")
co_air_df.head(1)
```

```python
alt.Chart(co_air_df).mark_line().encode(
    alt.X("month", title="Month"),
    alt.Y("aqi", title="Carbon Monoxide AQI"),
    color="year:N"
).properties(title="Carbon Monoxide")
```

```python
ozone_2019 = pd.read_csv("../../data/final/epa_air_params/ozone_2019_avg_df.csv")
ozone_2019_chart = alt.Chart(ozone_2019).mark_line().encode(
    alt.X("month", title="Month"),
    alt.Y("aqi", title="AQI"),
    color="year:N"
)
ozone_2019_chart
```

```python
ozone_2020 = pd.read_csv("../../data/final/epa_air_params/ozone_2020_avg_df.csv")
ozone_2020_chart = alt.Chart(ozone_2020).mark_line().encode(
    alt.X("month", title="Month"),
    alt.Y("aqi", title="AQI"),
    color="year:N"
)
ozone_2019_chart + ozone_2020_chart
```

```python
pm25_local_2019 = pd.read_csv("../../data/final/epa_air_params/pm25_local_2019_avg_df.csv")
pm25_local_2019_chart = alt.Chart(pm25_local_2019).mark_line().encode(
    alt.X("month", title="Month"),
    alt.Y("aqi", title="AQI"),
    color="year:N"
)
pm25_local_2019_chart
```

```python
pm25_local_2020 = pd.read_csv("../../data/final/epa_air_params/pm25_local_2020_avg_df.csv")
pm25_local_2020_chart = alt.Chart(pm25_local_2020).mark_line().encode(
    alt.X("month", title="Month"),
    alt.Y("aqi", title="AQI"),
    color="year:N"
)
(pm25_local_2019_chart + pm25_local_2020_chart).properties(title="PM2.5 - Local Conditions")
```

```python
pm25a_2019 = pd.read_csv("../../data/final/epa_air_params/pm25a_2019_avg_df.csv")
pm25a_2019_chart = alt.Chart(pm25a_2019).mark_line().encode(
    alt.X("month", title="Months"),
    alt.Y("aqi:Q", title="AQI"),
    color="year:N"
)
pm25a_2019_chart
```

```python
pm10_2019 = pd.read_csv("../../data/final/epa_air_params/pm10_2019_avg_df.csv")
pm10_2020 = pd.read_csv("../../data/final/epa_air_params/pm10_2020_avg_df.csv")

pm10_2019_chart = alt.Chart(pm10_2019).mark_line().encode(
    alt.X("month", title="Month"),
    alt.Y("aqi", title="AQI"),
    color="year:N"
)
pm10_2020_chart = alt.Chart(pm10_2020).mark_line().encode(
    alt.X("month", title="Month"),
    alt.Y("aqi", title="AQI"),
    color="year:N"
)
(pm10_2020_chart + pm10_2019_chart).properties(title="PM10 Total 0-10um STP")
```

```python
all_parm_2019 = pd.read_csv("../../data/final/epa_air_params/all_parms_2019_df.csv")
alt.Chart(all_parm_2019).mark_line().encode(
    alt.X("month", title="Month"),
    alt.Y("aqi", title="AQI"),
    color="parameter",
    strokeDash="parameter"
).properties(
    title="All Parameters 2019"
)
```

```python
all_parm_2020 = pd.read_csv("../../data/final/epa_air_params/all_parms_2020_df.csv")
alt.Chart(all_parm_2020).mark_line().encode(
    alt.X("month", title="Month"),
    alt.Y("aqi", title="AQI"),
    color="parameter",
    strokeDash="parameter"
).properties(
    title="All Parameters 2020"
)
```

```python
pm25_agg = pd.read_csv("../../data/final/epa_air_params/pm25_aggregated.csv")
alt.Chart(pm25_agg).mark_line().encode(
    alt.X("month(yearmonth):T", title="Month"),
    alt.Y("aqi:Q", title="AQI"),
    color="year:N",
    strokeDash="parameter"
)
```

```python
alt.Chart(pm25_agg).mark_rect().encode(
    alt.X("year:O", title="year"),
    alt.Y("month(yearmonth):O", title="month",),
    color="aqi:Q"
)
```

```python
a = alt.Chart(all_parm_2020).mark_rect().encode(
    alt.X("parameter"),
    alt.Y("month(date_month):O"),
    color="aqi:Q"
).properties(
    title="2020"
)
b = alt.Chart(all_parm_2019).mark_rect().encode(
    alt.X("parameter"),
    alt.Y("month(date_month):O"),
    color="aqi:Q"
).properties(
    title="2019"
)
alt.hconcat(b,a)
```
