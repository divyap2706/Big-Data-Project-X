```python
pm25a_2019_chart = alt.Chart(pm25a_2019).mark_line().encode(
    #alt.X("month"),
    alt.X("month"),
    alt.Y("aqi"),
    color="year:N"
)

pm25a_2020_chart = alt.Chart(pm25a_2020).mark_line().encode(
    #alt.X("month"),
    alt.X("month"),
    alt.Y("aqi"),
    color="year:N"
)

pm25a_chart = alt.layer(
    pm25a_2019_chart,
    pm25a_2020_chart
).properties(
    title="ACCEPTABLE PM2.5 AQI & SPECIATION MASS"
)
pm25a_chart
```

```python
pm25_local_2019_chart = alt.Chart(pm25_local_2019).mark_line().encode(
    alt.X("month"),
    alt.Y("aqi"),
    color="year:N"
).properties(
    title="PM2.5 Local Conditions - 2019"
)

pm25_local_2020_chart = alt.Chart(pm25_local_2020).mark_line().encode(
    alt.X("month"),
    alt.Y("aqi"),
    color="year:N"
).properties(
    title="PM2.5 Local Conditions - 2020"
)

pm25_local_chart = alt.layer(
    pm25_local_2019_chart,
    pm25_local_2020_chart
).properties(
    title="PM2.5 Local"
)
pm25_local_chart
```

```python
hotness_nyc_2020_chart = alt.Chart(avg_hotness_nyc_2020).mark_line().encode(
    alt.X("month"),
    alt.Y("avg(hotness_score)"),
    color="year:O",
    strokeDash="year:O"
).properties(
    title="NYC Real Estate Hotness"
)
hotness_metro_2020_chart = alt.Chart(avg_hotness_metro_2020).mark_line().encode(
    alt.X("month"),
    alt.Y("avg(hotness_score)"),
    color="year:O",
    strokeDash="year:O"
).properties(
    title="NYC Metro Real Estate Hotness"
)
# hotness_nyc_2020_chart
hotness_2020_chart = alt.layer(
    hotness_nyc_2020_chart,
    hotness_metro_2020_chart
)
hotness_2020_chart
```

```python
demand_nyc_2020_chart = alt.Chart(avg_hotness_nyc_2020).mark_line().encode(
    alt.X("month"),
    alt.Y("avg(demand_score)"),
    #color=alt.value("demand")
    #color="avg(supply_score)",
    #strokeDash="year:O"
).properties(
    title="NYC Real Estate Demand Score"
)
supply_nyc_2020_chart = alt.Chart(avg_hotness_nyc_2020).mark_line().encode(
    alt.X("month"),
    alt.Y("avg(supply_score)"),
    #color=alt.value("supply")
    #color="avg(supply_score)",
    #strokeDash="year:O"
).properties(
    title="NYC Real Estate Supply Score"
)
supply_demand_chart = alt.vconcat(
    supply_nyc_2020_chart,
    demand_nyc_2020_chart
)

alt.vconcat(
    pm25_local_chart,
    supply_demand_chart
#     alt.layer(
#         demand_nyc_2020_chart,
#         supply_nyc_2020_chart
#     )
)
#supply_nyc_2020_chart
```

```python
bridges_in_chart = alt.Chart(bridges_in_monthly).mark_line().encode(
    alt.X("month:Q", title="Month #"),
    alt.Y("Sum_Total_Vehicles:Q", title="# Vehicles"),
    color="Direction"
).properties(
    title="Vehicles Entering"
)
bridges_out_chart = alt.Chart(bridges_out_monthly).mark_line().encode(
    alt.X("month:Q", title="Month #"),
    alt.Y("Sum_Total_Vehicles:Q", title="# Vehicles"),
    color="Direction"
).properties(
    title="Vehicles Exiting via Bridge"
)

bridges_chart = alt.layer(
    bridges_in_chart,
    bridges_out_chart
)

alt.vconcat(
    pm25_local_chart,
    bridges_chart
)

```
