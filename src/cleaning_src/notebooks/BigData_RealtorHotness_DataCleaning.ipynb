```python
# Import the data and then use OpenClean's stream feature so that we do not need
# to put the entire file into memory
datafile = "/content/drive/MyDrive/Big Data/Big Data Project/Data Max is Working on/RDC_Inventory_Hotness_Metrics_Zip_History.csv"
ds = stream(datafile)

# Print the first ten rows of the dataset to get a first
# idea of the content.
ds.head()
```

```python
# Profile rows using the default data profiler.

from openclean.profiling.column import DefaultColumnProfiler

profiles = ds.profile(default_profiler=DefaultColumnProfiler)
```

```python
# Print overview of profiling results.

profiles.stats()
```

```python
#Print the distinct values of a column
print(ds.distinct('month_date_yyyymm'))
```

```python
# look at all the datatypes in the dataset
print(profiles.types())
```

```python
# query the minimum and maximum in the columns column
print(profiles.minmax('month_date_yyyymm'))
print(profiles.minmax('hotness_rank'))
print(profiles.minmax('hotness_rank_mm'))
print(profiles.minmax('hotness_rank_yy'))
print(profiles.minmax('hotness_score'))
```

```python
# Print the number of rows in the full dataset
print('\n{:,} rows.'.format(ds.count()))
```

```python
# Filter out dummy row with "quality_flag = 1:  year-over-year figures may be impacted"
from openclean.operator.transform.filter import filter
from openclean.function.eval.base import Col

filterOut = ds.filter(predicate=Col('month_date_yyyymm')!='quality_flag = 1:  year-over-year figures may be impacted')

# Recreate the profile and check if the string is gone from month_date_yyyymm and postal_code
profiles = filterOut.profile(default_profiler=DefaultColumnProfiler)
print(profiles.types())
```

```python
# Fix datatypes so that all numbers are floats
def to_float(in_num):
  if in_num == None or in_num == '':
    return ''
  elif type(in_num) != float:
    return float(in_num)
  else:
    return in_num

filterOut = filterOut.update('hotness_score', to_float)
filterOut = filterOut.update('median_listing_price_vs_us', to_float)
filterOut = filterOut.update('supply_score', to_float)
filterOut = filterOut.update('demand_score', to_float)
filterOut = filterOut.update('median_days_on_market', to_float)
filterOut = filterOut.update('median_dom_mm_day', to_float)
filterOut = filterOut.update('median_days_on_market_mm', to_float)
filterOut = filterOut.update('median_days_on_market_yy', to_float)
filterOut = filterOut.update('median_dom_yy_day', to_float)
filterOut = filterOut.update('median_dom_vs_us', to_float)
filterOut = filterOut.update('ldp_unique_viewers_per_property_mm', to_float)
filterOut = filterOut.update('ldp_unique_viewers_per_property_yy', to_float)
filterOut = filterOut.update('ldp_unique_viewers_per_property_vs_us', to_float)
filterOut = filterOut.update('median_listing_price', to_float)
filterOut = filterOut.update('median_listing_price_mm', to_float)
filterOut = filterOut.update('median_listing_price_yy', to_float)

profiles = filterOut.profile(default_profiler=DefaultColumnProfiler)
print(profiles.types())
```

```python
# query the minimum and maximum in the columns column
print(profiles.minmax('month_date_yyyymm'))
print(profiles.minmax('hotness_rank'))
print(profiles.minmax('hotness_score'))
print(profiles.minmax('supply_score'))
print(profiles.minmax('demand_score'))
```

```python
# Export data to file
# Note: Change the file path as needed
import pandas as pd

output = filterOut.to_df()
output.to_csv(r'/content/drive/MyDrive/Big Data/Big Data Project/Data Max is Working on/New_Realtor_Hotness.csv', index = False, header=True)

```
