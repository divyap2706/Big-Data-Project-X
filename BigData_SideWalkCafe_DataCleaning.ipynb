```python
# Import the data and then use OpenClean's stream feature so that we do not need
# to put the entire file into memory
datafile = "/content/drive/MyDrive/Big Data/Big Data Project/Data Max is Working on/Sidewalk_Caf__Licenses_and_Applications.csv"
ds = stream(datafile)

# Print the first ten rows of the dataset to get a first
# idea of the content.
ds.head()
```

```python
# Profile a sample of 1000 rows using the default data profiler.

from openclean.profiling.column import DefaultColumnProfiler

profiles = ds.profile(default_profiler=DefaultColumnProfiler)
```

```python
# Print overview of profiling results.

profiles.stats()
```

```python
#Print the distinct values of a column
print(ds.distinct('STATE'))
print(ds.distinct('CITY'))
print(ds.distinct('SWC_TYPE'))
```

```python
# look at all the datatypes in the dataset
print(profiles.types())
```

```python
# query the minimum and maximum in the columns column
print(profiles.minmax('APP_STATUS_DATE'))
print(profiles.minmax('EXPIRATION_DATE'))
print(profiles.minmax('APP_TOO_DATE'))
print(profiles.minmax('SUBMIT_DATE'))
print(profiles.minmax('SUBMIT_DATE'))
```

```python
# Print the number of rows in the full dataset
print('\n{:,} rows.'.format(ds.count()))
```

```python
# Update city column to be capitalized
from openclean.operator.transform.update import update
newds = ds.update('CITY', str.upper)
newds = newds.update('BUSINESS_NAME', str.upper)
newds = newds.update('BUSINESS_NAME2', str.upper)
print(newds.distinct('CITY'))
```

```python
# Define a function to fix the city misspellings
def city_transform(city):
  if city == 'LONG IS CITY':
    return 'LONG ISLAND CITY'
  else:
    return city

newerds = newds.update('CITY', city_transform)
print(newerds.distinct('CITY'))
```

```python
# Export data to file
# Note: Change the file path as needed
import pandas as pd

output = newerds.to_df()
output.to_csv(r'/content/drive/MyDrive/Big Data/Big Data Project/Data Max is Working on/New_Sidewalk_Cafe_Applications.csv', index = False, header=True)
```
