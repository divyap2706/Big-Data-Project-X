```python
# Import the data and then use OpenClean's stream feature so that we do not need
# to put the entire file into memory
datafile = "/content/drive/MyDrive/Big Data/Big Data Project/Data Max is Working on/RDC_Inventory_Core_Metrics_Zip_History.csv"
ds = stream(datafile)

# Print the first ten rows of the dataset to get a first
# idea of the content.
ds.head()
```

```python
# Profile rows using the default data profiler.

from openclean.profiling.column import DefaultColumnProfiler

profiles = ds.profile(default_profiler=DefaultColumnProfiler)
```

```python
# Print overview of profiling results.

profiles.stats()
```

```python
#Print the distinct values of a column
print(ds.distinct('month_date_yyyymm'))
```

```python
# look at all the datatypes in the dataset
print(profiles.types())
```

```python
# query the minimum and maximum in the columns column
print(profiles.minmax('month_date_yyyymm'))
print(profiles.minmax('median_listing_price'))
print(profiles.minmax('active_listing_count_yy'))
print(profiles.minmax('total_listing_count'))
```

```python
# Print the number of rows in the full dataset
print('\n{:,} rows.'.format(ds.count()))
```

```python
# Filter out dummy row with "quality_flag = 1:  year-over-year figures may be impacted"
from openclean.operator.transform.filter import filter
from openclean.function.eval.base import Col

filterOut = ds.filter(predicate=Col('month_date_yyyymm')!='* year-over-year figures may be impacted')

# Recreate the profile and check if the string is gone from month_date_yyyymm and postal_code
profiles = filterOut.profile(default_profiler=DefaultColumnProfiler)
print(profiles.types())
```

```python
# Fix datatypes so that all numbers are floats
def to_float(in_num):
  if in_num == None or in_num == '':
    return ''
  elif type(in_num) != float:
    return float(in_num)
  else:
    return in_num

filterOut = filterOut.update('median_listing_price', to_float)
filterOut = filterOut.update('median_listing_price_mm', to_float)
filterOut = filterOut.update('median_listing_price_yy', to_float)
#filterOut = filterOut.update('active_listing_count', to_float)
filterOut = filterOut.update('active_listing_count_mm', to_float)
filterOut = filterOut.update('active_listing_count_yy', to_float)
filterOut = filterOut.update('median_days_on_market', to_float)
filterOut = filterOut.update('median_days_on_market_mm', to_float)
filterOut = filterOut.update('median_days_on_market_yy', to_float)
#filterOut = filterOut.update('new_listing_count', to_float)
filterOut = filterOut.update('new_listing_count_mm', to_float)
filterOut = filterOut.update('new_listing_count_yy', to_float)
#filterOut = filterOut.update('price_increased_count', to_float)
filterOut = filterOut.update('price_increased_count_mm', to_float)
filterOut = filterOut.update('price_increased_count_yy', to_float)
#filterOut = filterOut.update('price_reduced_count', to_float)
filterOut = filterOut.update('price_reduced_count_mm', to_float)
filterOut = filterOut.update('price_reduced_count_yy', to_float)
#filterOut = filterOut.update('pending_listing_count', to_float)
filterOut = filterOut.update('pending_listing_count_mm', to_float)
filterOut = filterOut.update('pending_listing_count_yy', to_float)
filterOut = filterOut.update('median_listing_price_per_square_foot', to_float)
filterOut = filterOut.update('median_listing_price_per_square_foot_mm', to_float)
filterOut = filterOut.update('median_listing_price_per_square_foot_yy', to_float)
filterOut = filterOut.update('median_square_feet', to_float)
filterOut = filterOut.update('median_square_feet_mm', to_float)
filterOut = filterOut.update('median_square_feet_yy', to_float)
filterOut = filterOut.update('average_listing_price', to_float)
filterOut = filterOut.update('average_listing_price_mm', to_float)
filterOut = filterOut.update('average_listing_price_yy', to_float)
#filterOut = filterOut.update('total_listing_count', to_float)
filterOut = filterOut.update('total_listing_count_mm', to_float)
filterOut = filterOut.update('total_listing_count_yy', to_float)
filterOut = filterOut.update('pending_ratio', to_float)
filterOut = filterOut.update('pending_ratio_mm', to_float)
filterOut = filterOut.update('pending_ratio_yy', to_float)

profiles = filterOut.profile(default_profiler=DefaultColumnProfiler)
print(profiles.types())
```

```python
# query the minimum and maximum in the columns column
print(profiles.minmax('month_date_yyyymm'))
print(profiles.minmax('active_listing_count'))
print(profiles.minmax('new_listing_count'))
print(profiles.minmax('average_listing_price'))
```
