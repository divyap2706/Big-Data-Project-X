```python
!pip install git+git://github.com/VIDA-NYU/openclean.git
from openclean.pipeline import stream
from openclean.data.load import dataset
import pprint
import os

pp = pprint.PrettyPrinter(indent=2)
```

```python
# read the csv into the stream
datafile = "../../data/raw/filtered_ocean_and_estuary_data.csv"
ds = stream(datafile)
# ds = dataset(os.path.join("../../data/raw/", "ocean_and_estuary_data.csv"))

#Print the first ten rows of the dataset
ds.head()
```

```python
# profile the rows using the default profiler
from openclean.profiling.column import DefaultColumnProfiler

profiles = ds.profile(default_profiler=DefaultColumnProfiler)
```

```python
# Print overview of the profiling results
profiles.stats()
```

```python
# Print out columns with large numbers of empty values
for col in ds.columns:
    p = profiles.column(str(col))
    if p['emptyValueCount'] == 0:
        continue
    print(
        "\t'{}'\tEmpty: {:,},\tPercent Empty: {:.2%}".format(
            col, 
            p['emptyValueCount'], 
            p['emptyValueCount'] / p['totalValueCount']
        )
    )
```

```python
# Since there are multiple columns with no values, let's get rid of those
new_cols = []
print("Column Emptiness Stats")
for col in ds.columns:
    p = profiles.column(str(col))
    if p['emptyValueCount'] / p['totalValueCount'] < 0.5:
        new_cols.append(str(col))
        print(
            "\t'{}'\tEmpty: {:,},\tPercent Empty: {:.2%}".format(
                col, 
                p['emptyValueCount'], 
                p['emptyValueCount'] / p['totalValueCount']
            )
        )
ds = ds.select(new_cols)
print("\n\nColumnsRetained\n____________________________________________")
for col in new_cols:
    print(f"{col}")
```

```python
# check that we got the correct columns
profiles = ds.profile(default_profiler=DefaultColumnProfiler)

# Print out columns with large numbers of empty values
for col in ds.columns:
    p = profiles.column(str(col))
    if p['emptyValueCount'] == 0:
        continue
    print(
        "\t'{}'\tEmpty: {:,},\tPercent Empty: {:.2%}".format(
            col, 
            p['emptyValueCount'], 
            p['emptyValueCount'] / p['totalValueCount']
        )
    )
```
